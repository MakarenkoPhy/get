```python
import matplotlib.pyplot as plt
import numpy as np
Eps=0.0001
```

```python
def func1(x):
    return x**5 - x - 0.2
def func2(x):
    return 5.6 * np.sin(4.8 * x) - 4.5 * x
```

```python
def bisection_method(a, b, f):
    data=[]
    while(abs(a-b)>Eps):
        x0=(a + b)/2
        data.append([a, b, x0, f(a), f(b), f(x0)])
        if(f(a) * f(x0)) < 0:
            b=x0
        else:
            a=x0
    df=pd.DataFrame(data=data, columns=['a', 'b', 'x0', 'f(a)', 'f(b)', 'f(x0)'])
    return(np.round(x0, 3), df)
```

```python
def secant_method(a, b, f):
    data=[]
    x0=a-f(a) * (b-a) / (f(b) - f(a))
    while(abs(a-b)>Eps and abs(a-x0)>Eps and abs(b-x0)>Eps):
        data.append([a, b, x0, f(a), f(b), f(x0)])
        if(f(a) * f(x0)) < 0:
            b=x0
        else:
            a=x0
        x0=a-f(a) * (b-a) / (f(b) - f(a))
    df=pd.DataFrame(data=data, columns=['a', 'b', 'x0', 'f(a)', 'f(b)', 'f(x0)'])
    return(np.round(x0, 3), df)
```

```python
def func1_p(x):
    return 5 * x**4 - 1
def func2_p(x):
    return 5.6 * 4.8 * np.cos(4.8 * x) - 4.5
```

```python
def fi_func1(x):
    return x**5 - 0.2
def fi_func2(x):
    return 5.6 * np.sin(4.8*x)/4.5
```

```python
def newton_method(x, f, f_p, fi):
    data = [[x, fi(x)]]
    while(len(data) <= 3 or abs(data[-1][0] - data[-2][0]) / (1 - (data[-1][0] - data[-2][0]) / (data[-2][0] - data[-3][0])) > Eps ):
        x = x - f(x) / f_p(x)
        data.append([x, fi(x)])
    df=pd.DataFrame(data=data, columns=['x', 'fi(x)'])
    return(np.round(x, 3), df)
```

```python
def mod_newton_method(x, f, f_p, fi):
    x0=x
    data = []
    while(len(data) <= 3 or abs(data[-1][0] - data[-2][0]) / (1 - (data[-1][0] - data[-2][0]) / (data[-2][0] - data[-3][0])) > Eps ):
        data.append([x, fi(x)])
        x = x - f(x) / f_p(x0)
    df=pd.DataFrame(data=data, columns=['x', 'fi(x)'])
    return(np.round(x, 3), df)
```

```python
def sim(x, fi):
    data=[]
    while(len(data)<=2 or abs(data[-1][0] - data[-2][0]) > Eps):
        data.append([x, fi(x)])
        x = fi(x)
    df=pd.DataFrame(data=data, columns=['x', 'fi(x)'])
    return(np.round(x, 3), df)
```
